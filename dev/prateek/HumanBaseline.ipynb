```python
import pandas as pd
import numpy as np
from collections import Counter
pd.options.display.max_rows=100
```

```python
prateek_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Prateek Annotations.csv", skipinitialspace=True).replace(np.nan, '')
deep_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Deep Annotations.csv", skipinitialspace=True).replace(np.nan, '')
kartik_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Kartik Annotations.csv", skipinitialspace=True).replace(np.nan, '')
lakshita_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Lakshita Annotations.csv", skipinitialspace=True).replace(np.nan, '')
```

```python
all_labels = pd.read_pickle('/home/prateekagarw_umass_edu/p_file/all_labels')
all_labels['owner'] = all_labels['image_name'].str[0]
all_labels
```

```python
#Calculate Human Baseline 
top_1 = [ ]
top_3 = [ ]
groundtruth = all_labels['adjective_1'].tolist()
owner_list = all_labels['owner'].tolist()

for dfs in [(prateek_df,'p'),(deep_df,'d'),(lakshita_df,'l'),(kartik_df,'k')]:
    cur_df = dfs[0]
    cur_person = dfs[1]
    top_1_total = 0
    top_3_total = 0
    total_num = 0
    for idx,row in cur_df.iterrows():
        if cur_person == owner_list[idx]:
            continue
        total_num+=1
        adjectives = [row['adjective_1'],row['adjective_2'],row['adjective_3']]
        if groundtruth[idx] == adjectives[0]:
            top_1_total+=1
        if groundtruth[idx] in adjectives:
            top_3_total+=1
    top_1.append(top_1_total/total_num)
    top_3.append(top_3_total/total_num)
    
print(top_1)
print(top_3)
```

```python
#Plot the numbers
import matplotlib.pyplot as plt
top_1 = np.array(top_1)*100
top_3 = np.array(top_3)*100
std_1 = np.std(top_1)
mean_1 = np.mean(top_1)
std_3 = np.std(top_3)
mean_3 = np.mean(top_3)

print(f"The average top_1 accuracy is {mean_1} and the standard deviation is {std_1}")
print(f"The average top_3 accuracy is {mean_3} and the standard deviation is {std_3}")
```
