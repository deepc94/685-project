```python
import pandas as pd
import numpy as np
from collections import Counter
pd.options.display.max_rows=100
```

```python
prateek_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Prateek Annotations.csv", skipinitialspace=True).replace(np.nan, '')
deep_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Deep Annotations.csv", skipinitialspace=True).replace(np.nan, '')
kartik_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Kartik Annotations.csv", skipinitialspace=True).replace(np.nan, '')
lakshita_df = pd.read_csv("/home/prateekagarw_umass_edu/p_file/Lakshita Annotations.csv", skipinitialspace=True).replace(np.nan, '')
```

```python
#Calculate Top-1 Agreement
agreement = 0
for values in zip(prateek_df['adjective_1'].tolist(),lakshita_df['adjective_1'].to_list(),deep_df['adjective_1'].tolist(),kartik_df['adjective_1'].tolist()):
    if (values[0] == values[1] == values[2] == values[3]):
        agreement+=1
    else:
        continue

print(agreement)
print(agreement/257)
```

```python
#Locate the Negatives and add them to the all_aug_dataset
negative_adjectives = []

for values in zip(prateek_df['negative_adj'].tolist(),lakshita_df['negative_adj'].to_list(),deep_df['negative_adj'].tolist(),kartik_df['negative_adj'].tolist()):
    #Create the negative set
    negative_set = set({values[0],values[1],values[2],values[3]})
    if '' in negative_set:
        negative_set.remove('')
    negative_adjectives.append(list(negative_set))
    
# print(negative_adjectives)
```

```python
all_labels_aug = pd.read_csv("/home/prateekagarw_umass_edu/p_file/all_labels_aug.csv", skipinitialspace=True)
all_labels_aug['image_name'] = all_labels_aug['image_name'] + ".jpg"
all_labels_aug['negative_adj'] = negative_adjectives
all_labels_aug['object_1'] = prateek_df['object_1']
all_labels_aug['object_2'] = prateek_df['object_2']
all_labels_aug 
```
